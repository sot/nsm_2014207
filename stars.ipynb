### Target (expected) attitude for acquisition

```python
att = Quat([0.56933254, 0.02908272, -0.37884253, 0.72903566])
```

```python
att.ra, att.dec, att.roll
```

### AGASC stars within 1.5 deg of target attitude

```python
SKA = os.environ.get('SKA', os.path.join(os.environ['HOME'], 'ska'))  # SKA or ~/ska
agasc_file = os.path.join(SKA, 'data', 'agasc', 'miniagasc.h5')
stars_all = agasc.get_agasc_cone(att.ra, att.dec, agasc_file=agasc_file, radius=1.5)
```

```python
len(stars_all)
```

#### Convert to an astropy table

```python
stars = Table(stars_all)
```

#### Convert to `yag`, `zag` for target attitude

```python
yag, zag = Ska.quatutil.radec2yagzag(stars['RA_PMCORR'], stars['DEC_PMCORR'], att)
stars['yag'] = yag * 3600
stars['zag'] = zag * 3600
```

#### Get the distance in arcsec between the two acquired stars:

```python
osy = obs_stars['yag']
osz = obs_stars['zag']
dist12 = np.sqrt((osy[1] - osy[0])**2 + (osz[1] - osz[0])**2)
dist12
```

### Find star pairs that match the relative offsets of acquired stars

This is an easily-understood but inefficient implementation.  Loop through every pair of stars in `catalog` and look for:

- Pair distance within 20 arcsec of `dist12`
- Absolute delta y and delta z distance between pair similar to the acquisition pair
- Consider only one-shot of up to 400 arcsec 

```python
dists = []
yag = stars['yag']
zag = stars['zag']
# Pick catalog stars within a 400 arcsec half-width box of either acq star
ok = (((np.abs(yag - osy[0]) < 400) | (np.abs(yag - osy[1]) < 400))
      & ((np.abs(zag - osz[0]) < 400) | (np.abs(zag - osz[1]) < 400)))
sok = stars[ok]
yok = yag[ok]
zok = zag[ok]

# Check all candidate pairs in catalog and store possible matches
for y1, z1, s1 in zip(yok, zok, sok):
    for y2, z2, s2 in zip(yok, zok, sok):
        dist = np.sqrt((y1 - y2)**2 + (z1 - z2)**2)
        dy0 = y1 - osy[0]
        dy1 = y2 - osy[1]
        dz0 = z1 - osz[0]
        dz1 = z2 - osz[1]
        if abs(dist - dist12) < 10 and abs(dy0 - dy1) < 20 and abs(dz0 - dz1) < 20:
            dists.append({'ddist': dist - dist12, 
                          'dy0': dy0, 'dy1': dy1, 
                          'dz0': dz0, 'dz1': dz1, 
                          'mag1': s1['MAG_ACA'], 'mag2': s2['MAG_ACA'],
                          'id1': s1['AGASC_ID'], 'id2': s2['AGASC_ID']})
```

**Only one star pair found:**

```python
dists
```

```python
def make_plot(xlim=(2600, -2600), ylim=(-2600, 2600)):
    plt.figure(figsize=(5, 5))
    plt.plot(yag, zag, '.', markersize=3)
    plt.xlim(*xlim)
    plt.ylim(*ylim)

    # Observed stars in RED assuming target attitude (zero manuever error)
    plt.plot(obs_stars['yag'], obs_stars['zag'], '.r', markersize=10)

    # Bright stars in CYAN
    ok = stars['MAG_ACA'] < 10.5
    plt.plot(yag[ok], zag[ok], '.c', markersize=10, alpha=0.5)

    # Candidate identifications for observed stars in BLUE
    ok = stars['AGASC_ID'] == 910689112
    plt.plot(yag[ok], zag[ok], '.b', markersize=10, alpha=0.5)
    ok = stars['AGASC_ID'] == 910689744
    plt.plot(yag[ok], zag[ok], '.b', markersize=10, alpha=0.5)
    
    plt.grid()
    plt.xlabel('Y angle (arcsec)')
    plt.ylabel('Z angle (arcsec)')
```

```python
make_plot()
```
